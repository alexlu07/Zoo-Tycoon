```python
import numpy as np
import cv2
```

```python
def show(img, size=1000):
    h, w = img.shape[:2]
    m = max(h, w)

    img = cv2.resize(img, (int(w/m*size), int(h/m*size)))
    cv2.imshow("image", img)
    cv2.waitKey(0) 
    cv2.destroyAllWindows() 
```

```python
im = cv2.imread('assets/images/elephant.png')

imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 127, 255, 0)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

blank_img = np.zeros_like(im)
outline = cv2.drawContours(blank_img, contours[:3] + contours[4:], -1, (0, 255, 0), 3)
```
